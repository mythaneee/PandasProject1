```python
""" Extracts all Legendary Pokemon into a new DataFrame. """

import pandas as pd
df = pd.read_csv("FirstGenPokemon.csv")  

legendaries = df[df["Legendary"] == True]

print(legendaries.to_string()) 

""" Technical discovery: This demonstrates how we can filter a DataFrame based on its class level for comparison or further analysis. """
```

```python
""" Find all "Fire" types (checks both Type1 and Type2). """

import pandas as pd
df = pd.read_csv("FirstGenPokemon.csv")  

fire_pokemon = df[(df["Type1"] == "Fire") | (df["Type2"] == "Fire")]

print(fire_pokemon.to_string())

""" Technical discovery: This shows how we can use boolean indexing to filter rows based on multiple conditions, allowing us to find all Pokemon of a specific type regardless of their primary or secondary classification. """
```

```python
""" Filters Water type Pokemons with Weight > 100kg. """

import pandas as pd
df = pd.read_csv("FirstGenPokemon.csv")  

heavy_water_pokemon = df[((df["Type1"] == "Water") | (df["Type2"] == "Water")) & (df["Weight"] > 100)]

print(heavy_water_pokemon.to_string())

""" Technical discovery: This demonstrates how we can incorporate multiple conditions for multi-criteria feature engineering. """
```

```python
""" Quantifies missing values per column using isna().sum() """

import pandas as pd
df = pd.read_csv("FirstGenPokemon.csv")  

print(df.isna().sum())

""" Technical discovery: This demonstrates how we can quickly identify columns with missing values, which is crucial for data cleaning and preprocessing. """
```

```python
""" Replaces NaN values in Type2 with "None" and prints verification. """

import pandas as pd
df = pd.read_csv("FirstGenPokemon.csv")  

none_in_type2 = df.fillna({"Type2": "None"})

print(none_in_type2.to_string())

""" Technical discovery: This demonstrates how we can make downstream grouping/printing clearer by replacing NaN values with a more descriptive placeholder, which can help avoid confusion during analysis. """
```

```python
""" Casts the "Legendary" column to boolean type. """

import pandas as pd
df = pd.read_csv("FirstGenPokemon.csv")  

df["Legendary"] = df["Legendary"].astype(bool)

print(df.to_string())

""" Technical discovery: This is an example of how we can ensure that later filters and group-bys behave consistently. """
```

```python
""" Applies .str.title() ensuring naming consistency. """

import pandas as pd
df = pd.read_csv("FirstGenPokemon.csv")  

df["Name"] = df["Name"].str.title()

print(df.to_string())

""" Technical discovery: Standardizing string formats can help ensure consistency across the dataset, which is important for accurate analysis and presentation. """
```

```python
""" Calculates the global means for both "Height" and "Weight". """

import pandas as pd
df = pd.read_csv("FirstGenPokemon.csv")  

print(df[["Height", "Weight"]].mean(numeric_only=True))

""" Technical discovery: This is ideal for continuous variables for comparison between before per-type or per-group analysis. """
```

```python
""" Identifies specific name of pokemon that is the tallest and heaviest. """

import pandas as pd
df = pd.read_csv("FirstGenPokemon.csv")

tallest_and_heaviest = df[["Height", "Weight"]].max(numeric_only=True)

# Use '&' so only rows matching both max Height and max Weight are returned.
result = df[(df["Height"] == tallest_and_heaviest["Height"]) & (df["Weight"] == tallest_and_heaviest["Weight"])]

print(result[["Name", "Height", "Weight"]])

""" Technical discovery: Useful for identifying outliers """
```

```python
""" Computes frequency distribution of Pokemon types (Type1). """

import pandas as pd
df = pd.read_csv("FirstGenPokemon.csv")


print(df["Type1"].value_counts())

""" Technical discovery: This is a common way to understand the distribution of categorical variables, which can inform further analysis or visualization. """
```

```python
""" Groups Pokemon by Type1 and computes average Height and Weight for each type. """

import pandas as pd
df = pd.read_csv("FirstGenPokemon.csv")

group = df.groupby("Type1")[["Height", "Weight"]].mean(numeric_only=True)

print(group.to_string())

""" Technical discovery: Enables trait comparison across groups, which can reveal interesting patterns or differences between types. """
```

```python
""" Determines which Pokemon type (Type1) has the highest average weight. """

import pandas as pd
df = pd.read_csv("FirstGenPokemon.csv")

max_weight = df.groupby("Type1")["Weight"].mean(numeric_only=True).idxmax()

print(max_weight)
```

```python
""" Contrasts Legendary vs non-Legendary Pokemon by computing count, average Height and Weight for each group. """

import pandas as pd
df = pd.read_csv("FirstGenPokemon.csv")

legendary_stats = df.groupby("Legendary")[["Height", "Weight"]].agg(["count", "mean"])
print(legendary_stats.to_string().title())

""" Technical discovery: Reveals distribution and scale differences between cohorts in one compact grouped table. """
```

```python
""" Among non-Legendary Pokemon, identifies the heaviest Type1 group. """

import pandas as pd
df = pd.read_csv("FirstGenPokemon.csv")

non_legendary = df[df["Legendary"] == 0]

heaviest_type = non_legendary.groupby("Type1")["Weight"].max().idxmax()

print(heaviest_type)

```

```python
""" Anayzes dual-type combinations and finds the Top 5 most frequent pairings. """

import pandas as pd
df = pd.read_csv("FirstGenPokemon.csv")

type_combinations = df.groupby(["Type1", "Type2"]).size().reset_index(name="Count")
top_combinations = type_combinations.sort_values(by="Count", ascending=False).head(5)

print(top_combinations.to_string(index=False))
```

```python
""" Bonus Challenge: Finds the lightest Flying-type Pokemon. """

import pandas as pd
df = pd.read_csv("FirstGenPokemon.csv")

flying_pokemon = df[(df["Type1"] == "Flying") | (df["Type2"] == "Flying")]
lightest_idx = flying_pokemon["Weight"].idxmin()
lightest_flying = flying_pokemon.loc[lightest_idx, ["Name", "Type1", "Type2", "Weight"]]

print(lightest_flying.to_string())
```
